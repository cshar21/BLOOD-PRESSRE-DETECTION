```python
import numpy as np
import scipy.io
import os

# Segmentation function (you already used this)
def segment_signals(ppg, abp, window_size=256, step_size=128):
    segments_ppg = []
    segments_abp = []
    for i in range(0, len(ppg) - window_size + 1, step_size):
        segments_ppg.append(ppg[i:i + window_size])
        segments_abp.append(abp[i:i + window_size])
    return segments_ppg, segments_abp
```

```python
from tqdm import tqdm

mat_files = sorted([f for f in os.listdir(local_dataset_path) if f.endswith('.mat')])

batch_size = 2  # Adjust to avoid crash (2–3 files is safe for Colab)
num_batches = len(mat_files) // batch_size + (len(mat_files) % batch_size != 0)

for batch_idx in range(num_batches):
    batch_files = mat_files[batch_idx * batch_size : (batch_idx + 1) * batch_size]

    all_ppg_segments = []
    all_abp_segments = []

    print(f"\n🔄 Processing Batch {batch_idx+1}/{num_batches} — Files: {batch_files}")

    for fname in batch_files:
        print(f"📂 Loading {fname}...")
        fpath = os.path.join(local_dataset_path, fname)
        mat = scipy.io.loadmat(fpath)
        records = mat['p'].squeeze()

        for sample in records:
            ppg, abp = sample[0], sample[1]
            seg_ppg, seg_abp = segment_signals(ppg, abp)
            all_ppg_segments.extend(seg_ppg)
            all_abp_segments.extend(seg_abp)

    # Convert to NumPy arrays
    X = np.array(all_ppg_segments)
    y = np.array(all_abp_segments)

    # Optional normalization (use only if you want)
    # X = np.array([(x - np.min(x)) / (np.max(x) - np.min(x) + 1e-8) for x in X])
    # y = np.array([(a - np.min(a)) / (np.max(a) - np.min(a) + 1e-8) for a in y])

    # Save this batch to disk
    save_path = f'/content/ppg_abp_batch_{batch_idx+1:02d}.npz'
    np.savez_compressed(save_path, X=X, y=y)
    print(f"✅ Saved {X.shape[0]} segments to {save_path}")
```

```python
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error

svr_model = SVR(kernel='rbf', C=1.0, epsilon=0.1)

for i, fpath in enumerate(batch_paths):
    print(f"🔄 Batch {i+1}/{len(batch_paths)}: {fpath}")

    data = np.load(fpath)
    X, y = data['X'], data['y']

    # Use mean ABP as label
    y_mean = y.mean(axis=1)

    # Sample first 10k (adjust based on time/memory)
    X_sample = X[:10000]
    y_sample = y_mean[:10000]

    svr_model.fit(X_sample, y_sample)
    print(f"✅ Trained on {len(y_sample)} samples")
```

```python
# ✅ CNN model building and training with batch-wise input

import os
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models

# --------------------------
# Step 1: Build CNN Model
# --------------------------
def build_cnn_model():
    model = models.Sequential([
        layers.Input(shape=(256, 1)),
        layers.Conv1D(32, 5, activation='relu'),
        layers.MaxPooling1D(2),
        layers.Conv1D(64, 5, activation='relu'),
        layers.MaxPooling1D(2),
        layers.Conv1D(128, 3, activation='relu'),
        layers.GlobalAveragePooling1D(),
        layers.Dense(256, activation='linear')  # output shape
    ])
    model.compile(optimizer='adam', loss='mse')
    return model

cnn_model = build_cnn_model()

# --------------------------
# Step 2: Batch-wise Training
# --------------------------
# Find all batch files
batch_files = sorted([
    f for f in os.listdir("/content")
    if f.startswith("ppg_abp_batch") and f.endswith(".npz")
])

# Loop through batches and train
for i, file in enumerate(batch_files):
    print(f"\n🔄 CNN Training — Batch {i+1}/{len(batch_files)}: {file}")

    # Load batch data
    data = np.load(os.path.join("/content", file))

    # Replace these keys based on how you saved the files
    X_batch = data['X']
    y_batch = data['y']

    # Normalize PPG input (optional but useful)
    X_batch = X_batch / np.max(X_batch)
    X_batch = X_batch[..., np.newaxis]  # Reshape for Conv1D: (samples, 256, 1)

    # Subsample if needed to fit memory
    sample_size = min(10000, len(X_batch))
    idx = np.random.choice(len(X_batch), sample_size, replace=False)

    # Train model
    cnn_model.fit(
        X_batch[idx],
        y_batch[idx],
        epochs=3,
        batch_size=64,
        verbose=1
    )
```

```python
import numpy as np
import os
from sklearn.model_selection import train_test_split

def load_npz_data(path='/content/batches', limit=None):
    X, Y = [], []
    for f in sorted(os.listdir(path)):
        if f.endswith('.npz'):
            data = np.load(os.path.join(path, f))
            x, y = data['x'], data['y']
            X.append(x)
            Y.append(y)
            if limit and len(X) * len(x) >= limit:
                break
    X = np.concatenate(X)
    Y = np.concatenate(Y)
    X = (X - X.mean()) / X.std()  # normalize
    return train_test_split(X, Y, test_size=0.2, random_state=42)

X_train, X_test, y_train, y_test = load_npz_data('/content/batches', limit=20000)  # limit to speed up
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

rf = RandomForestRegressor(n_estimators=100, n_jobs=-1)
rf.fit(X_train, y_train)
pred_rf = rf.predict(X_test)

print("RF MSE:", mean_squared_error(y_test, pred_rf))
```
